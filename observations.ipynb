```python
import re

import requests
from bs4 import BeautifulSoup
```

```python
observations = {}
counter = 0
while(True):
    print(counter)
    response = requests.post(
        'https://www.enquetes-publiques.com/Enquetes_WEB/FR/CONSULTER-MDP4.awp?P1=EP17491',
        params={
            'P1': 'EP17491',
            'AWPIDEEA696BA': '030ED7D5ACFE5E5A7C75C3E42267B233541173AE',
        },
        headers={
            'Cookie': 'AWP_CSESSIONEEA696BA=030ED7D5ACFE5E5A7C75C3E42267B233541173AE',
            'Referer': 'https://www.enquetes-publiques.com/Enquetes_WEB/FR/CONSULTER-MDP4.awp?P1=EP17491',
            'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Ubuntu Chromium/65.0.3325.181 Chrome/65.0.3325.181 Safari/537.36',
        },
        data={
            'WD_ACTION_': 'AJAXEXECUTE',
            'SCROLLTABLE': 'A40',
            'A40': str(counter),
        },
    )
    assert response.status_code == 200
    
    starts = [
        m.end()
        for m in re.finditer('\[CDATA\[', response.text)
    ]
    ends = [
        m.start()
        for m in re.finditer('\]\]', response.text)
    ]
    chunks = [
        response.text[start:end]
        for start, end in zip(starts, ends)
    ]
    if not chunks:
        break

    for chunk in chunks:
        soup = BeautifulSoup(chunk, 'html.parser')
        observations_raw = soup.find_all(class_='pos54')
        for observation_raw in observations_raw:
            index = int(observation_raw.find(class_='pos56').find('table').find('tr').find('td').find('div').text)
            text = observation_raw.find(class_='pos58').find('table').find('tr').find('td').find('div').get_text(separator='\n')
            observations[index] = text

    counter += 5
```

```python
observations_list = [
    observations[i+1]
    for i in range(len(observations))
]
```

```python
with open('observations.txt', 'w') as f:
    for i in range(len(observations_list)):
        f.write('------- {} -------\n'.format(i+1))
        f.write(observations_list[i])
        f.write('\n')
        f.write('\n')
```
